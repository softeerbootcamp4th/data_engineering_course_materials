## # gear, transmission - crosstab

````python
pd.crosstab(mtcars_df['gear'], mtcars_df['am'])
````

````python
# gear, am의 경우의 수 조합별로 나누고 각 조합에 몇개의 그룹이 있는지
# pandas.crosstab method 이용!
mtcars_df_crosstab_gear_am = pd.crosstab(mtcars_df['gear'], mtcars_df['am'])

import matplotlib.pyplot as plt
# 인덱스와 컬럼으로 crosstab을 뽑아줌
# mtcars_df_crosstab_gear_am.index => gear
# mtcars_df_crosstab_gear_am.column => am
# 조합 찾기 => x축
# y는 값
x = []
y = []
for gear_idx, gear in enumerate(mtcars_df_crosstab_gear_am.index):
    for am_idx, am in enumerate(mtcars_df_crosstab_gear_am.columns):
        # 없는 값은 안그릴거야
        if mtcars_df_crosstab_gear_am.iloc[gear_idx, am_idx] == 0:
            continue
        x.append(str((gear, am)))
        y.append(mtcars_df_crosstab_gear_am.iloc[gear_idx, am_idx])

# 그림 그리기
plt.bar(x,y)

plt.title('# of Cars by Gear x Transmission')
plt.xlabel('(Gear, Auto Transmission)')
plt.ylabel('# of Cars')

plt.show()
````

## # gear, transmission - groupby

````python
# gear, am의 경우의 수 조합별로 나누고 각 조합에 몇개의 그룹이 있는지
# => groupby -> size, reset_index
# 결과값 출력 먼저
mtcars_df_groupby_gear_am = mtcars_df.groupby(['gear', 'am']).size().reset_index(name='counts')
print(mtcars_df_groupby_gear_am)

# 바 그래프


# x축 만들기
# iloc => (Gear, Auto Transmission) 튜플 리스트 생성
x = [str(tuple(mtcars_df_groupby_gear_am.iloc[i][0:2])) for i in range(len(mtcars_df_groupby_gear_am))]
#print(x)
# y축
y = list(mtcars_df_groupby_gear_am['counts'])
#print(y)
# 그리기
plt.bar(x, y)

plt.title('# of Cars by Gear x Transmission')
plt.xlabel('(Gear, Auto Transmission)')
plt.ylabel('# of Cars')

plt.show()
````

- 변수들 간의 상관관계가 높은 조합을 임의로 2개 선택
- 해당 데이터 간의 상관관계를 그래프로 그리자 !
- 어떤 결론을 내릴 수 있나?

````python
# cyl - disp => 0.9
# disp - wt => 0.89
# disp가 다른 피처에 영향을 많이 미치는 요소일수도? 
# 2개만 봐서는 잘 모르겠다
# 그래프를 그려보자 !
````

````python
import networkx as nx
import numpy as np
from collections import defaultdict
def plot_network_graph(correlation_matrix, threshold=0.7):
    # 상관계수 매트릭스 => 인접행렬
    adj_matrix = np.abs(correlation_matrix)  # 절대값

    # threshold 값 미만 다 0 => 간선 안그림
    adj_matrix[adj_matrix < threshold] = 0
    adj_matrix[adj_matrix == 1] = 0

    # 네트워크 그래프 생성
    G = nx.from_numpy_array(adj_matrix.values)
    # 노드 생성
    G = nx.relabel_nodes(G, lambda x: correlation_matrix.columns[x])

    # 그래프 그리기
    plt.figure(figsize=(8, 6))
    pos = nx.spring_layout(G)
    nx.draw(G, pos, with_labels=True, node_size=0, node_color='skyblue', font_size=15, font_weight='bold', edge_color='gray')

    # 엣지 레이블 추가 (상관계수)
    edge_labels = defaultdict(str)
    for i in range(correlation_matrix.shape[0]):
        for j in range(correlation_matrix.shape[1]):
            if adj_matrix.iloc[i, j] > threshold:
                edge_labels[(correlation_matrix.index[i], correlation_matrix.columns[j])] = f'{adj_matrix.iloc[i, j]:.2f}'
    nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_size=12)

    plt.title("Correlation Network Graph")
    plt.show()

plot_network_graph(correlation_matrix)
````

````python
# 상관계수가 0.7 이상인 간선만을 표시
# cyl이 다른 변수들과 높은 상관관계를 많이 가지는 것으로 보임
# cyl값의 변화가 다른 피처의 변화에 많은 영향을 미칠 수 있을 것 같음 !
````

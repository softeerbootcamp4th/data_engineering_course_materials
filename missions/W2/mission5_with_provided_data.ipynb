```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from wordcloud import WordCloud
import itertools
from collections import defaultdict, Counter
# pd.read_csv('/Users/admin/Desktop/HMG/ws/Mission/W2/data.csv')
data = pd.read_csv('/Users/admin/Desktop/HMG/ws/Mission/W2/data.csv', sep=',', encoding='latin-1', on_bad_lines='skip')
data.columns = ['sentiment', 'ID', 'timestamp', 'query_flag', 'username', 'text']
```

```python
positive_words = ' '.join(data.text.loc[data.sentiment==4].sample(frac=0.1).to_list())
negative_words = ' '.join(data.text.loc[data.sentiment==0].sample(frac=0.1).to_list())
```

```python
positive_wc = WordCloud(background_color="white", max_words=200)
positive_cloud = positive_wc.generate(positive_words)
negative_wc = WordCloud(background_color="white", max_words=200)
negative_cloud = negative_wc.generate(negative_words)
```

```python
plt.figure(figsize=(10,10))

plt.subplot(211) # comma 생략가능
plt.xticks([], [])
plt.yticks([], [])
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
plt.gca().spines['bottom'].set_visible(False)
plt.gca().spines['left'].set_visible(False)
plt.imshow(positive_cloud)
plt.title('Positive Tweets Word Cloud')

plt.subplot(212) # comma 생략가능
plt.xticks([], [])
plt.yticks([], [])
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
plt.gca().spines['bottom'].set_visible(False)
plt.gca().spines['left'].set_visible(False)
plt.imshow(negative_cloud)
plt.title('Negative Tweets Word Cloud')
```

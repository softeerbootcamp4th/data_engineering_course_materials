### 전체 제목 Word Cloud

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
from konlpy.tag import Okt
from collections import Counter
```

```python
def make_word_cloud(arg, description):
    okt = Okt()
    nouns = okt.nouns(arg)  # 명사만 추출
    words = [n for n in nouns if len(n) > 1]  # 단어의 길이가 1개인 것은 제외
    cons = Counter(words)  # 단어별 빈도수 딕셔너리 데이터 생성

    # 특정 단어와 그 횟수 제거하기
    words_to_remove = ["회사", "직원"]
    for remove_word in words_to_remove:
        if remove_word in cons:
            del cons[remove_word]

    # 특정 단어 교체하기
    if "라벨" in cons:
        cons["워라벨"] = cons.pop("라벨")

    # font_path = "/usr/share/fonts/truetype/nanum/NanumGothic.ttf"
    # sys_font = fm.findSystemFonts()
    # [f for f in sys_font if "nanum" in f]
    # img = Image.open("/Users/admin/Documents/GitHub/softeer_wiki/missions/w2/logo.png")
    # img_array = np.array(img)
    wc = WordCloud(
        font_path="AppleGothic",
        width=400,
        height=400,
        scale=2.0,
        max_font_size=250,
        background_color="white",
        # mask=img_array,
        random_state=42,  # 동일한 결과를 얻기 위해 random_state 설정
    )
    result = wc.generate_from_frequencies(cons)
    plt.figure()
    plt.title(description)  # 제목 추가
    plt.imshow(result, interpolation="bilinear")
    plt.axis("off")
    plt.show()
    return result
```

```python
# 모든 제목을 연결한 텍스트
df0 = df["제목"].str.cat(sep=" ")

df1 = df["장점"].str.cat(sep=" ")

df2 = df["단점"].str.cat(sep=" ")

make_word_cloud(df0, "제목 워드 클라우드")

make_word_cloud(df1, "장점 워드 클라우드")

make_word_cloud(df2, "딘점 워드 클라우드")
```

```python
df3 = df[df["총점"] < 2]["제목"].str.cat(sep=" ")
make_word_cloud(df3, "총점이 1점일때 제목 워드 믈라우드")
```

```python
df3 = df[df["총점"] > 4]["제목"].str.cat(sep=" ")
make_word_cloud(df3, "총점이 5점일때 제목 워드 믈라우드")
```

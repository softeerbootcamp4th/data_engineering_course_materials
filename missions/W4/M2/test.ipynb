## 현재 작업중인 컬럼

````python
# 모든 값이 datetime64[us]인지 확인하는 함수
def is_datetime64_us(value):
    if pd.isnull(value):
        return False
    if not isinstance(value, pd.Timestamp):
        return False
    if value.nanosecond % 1000 != 0:  # 마이크로초가 아닌 값이 있는지 확인
        return False
    return True

# 각 행에 대해 함수 적용하여 Boolean 값 생성
df['checked_request_datetime'] = df['request_datetime'].apply(is_datetime64_us)

# Boolean 값이 False인 행만 출력
df[df['checked_request_datetime'] == False]

# Boolean 값이 False인 행을 제거
df = df[df['checked_request_datetime'] == True]
````

## 현재 작업중인 컬럼
* PULocationID 
* TLC Trip Record Data에 따르면 PULocationID의 값은 1 ~ 265인 정수이다.
* TLC Trip Record Data가 바뀌면 유효값 범위가 바뀔 수 있으나 고려하지 않는다.

````python
# 모든 값이 1 ~ 265 사이의 정수인지 확인하는 함수
def is_int_between_1_and_265(value):
    if pd.isnull(value):
        return False
    if not isinstance(value, int):
        return False
    if value < 1 or value > 265:
        return False
    return True

# 각 행에 대해 함수 적용하여 Boolean 값 생성
df['checked_PULocationID'] = df['PULocationID'].apply(is_int_between_1_and_265)

# Boolean 값이 False인 행만 출력
df[df['checked_PULocationID'] == False]

# Boolean 값이 False인 행을 제거
df = df[df['checked_PULocationID'] == True]
````

````python
# 시간 정보를 추출
df['hour'] = df['pickup_datetime'].dt.hour

# 시간별로 그룹화하고 개수 세기
hour_counts = df.groupby('hour').size()

# 결과를 데이터프레임으로 변환
hour_counts_df = hour_counts.reset_index(name='count')

# 결과 확인
print(hour_counts_df)
````

### Visualize the distribution of trips across different hours of the day.

````python
import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter

# 최대 값을 가진 행의 인덱스를 찾기
max_index = hour_counts_df['count'].idxmax()

# 막대 색 설정 (기본 색: blue, 최대 값: red)
colors = ['blue' if i != max_index else 'red' for i in range(len(hour_counts_df))]

# y축 레이블을 만명 단위로 포맷하는 함수 정의
def thousands(x, pos):
    return '%1.0f' % (x * 1e-4)

# 막대 그래프 그리기
plt.figure(figsize=(12, 6))
plt.bar(hour_counts_df['hour'], hour_counts_df['count'], color=colors)
plt.xlabel('Hour')
plt.ylabel('Count (10,000 people)')
plt.title('Call Count by Hour')

# y축 포맷터 설정
formatter = FuncFormatter(thousands)
plt.gca().yaxis.set_major_formatter(formatter)

plt.show()
````

````python
# 상관 계수 계산
correlation = hour_counts_df['hour'].corr(hour_counts_df['count'])
print(f'Correlation between hour and taxi rides count: {correlation:.2f}')
````

## Weather Condition Analysis
### Analyze the effect of weather conditions on taxi demand. Use additional datasets if necessary to obtain weather information for the corresponding time periods.



````python
# 날씨 데이터를 df_weather에 저장
"""
date,hour,Time,Temperature (F),Condition
"""
df_weather = pd.read_csv('weather_.csv')

# df_weather의 hour 컬럼을 int32로 변환
df_weather['hour'] = df_weather['hour'].astype('int32')

# df_weather의 Time 컬럼을 drop
df_weather = df_weather.drop(columns=['Time'])

# df의 pickup_datetime 에서 날짜를 추출하여 date 컬럼을 생성    
df['date'] = df['pickup_datetime'].dt.date
df



# df의 타입을 확인
print(df.dtypes)

# df_weather의 타입을 확인
print(df_weather.dtypes)
````

````python
# df의 date 컬럼의 타입을 str로 변환
df['date'] = df['date'].astype('str')

# df의 request_datetime, pickup_datetime, dropoff_datetime, PULocationID, DOLocationID, trip_miles, trip_time 컬럼을 drop
df = df.drop(columns=['request_datetime', 'pickup_datetime', 'dropoff_datetime', 'PULocationID', 'DOLocationID', 'trip_miles', 'trip_time'])

# date, hour 컬럼을 기준으로 groupby 하고 count 컬럼 추가
df = df.groupby(['date', 'hour']).size().reset_index(name='count')

# df 와 df_weather을 date, hour 컬럼을 기준으로 left join
df = pd.merge(df, df_weather, on=['date', 'hour'], how='left')
````

어떤 데이터가 주어졌을 때, 독립 변수들은 대부분 상관 관계를 가지고 있다. 이로 인한 문제를 '다중 공선성 문제'라고 하며 이런 문제를 해결하기 위해 독립 변수들에 적절한 조치를 취해야한다. 또한 종속변수를 정할 때, 즉 구하고자 하는 것을 명확히 하여 적절한 종속 변수를 설정해야한다.

1. 구하고자 하는 것 : 날씨에 따른 택시 수요
'택시 수요'는 종속 변수, 날씨는 독립 변수이다. '택시의 수요'는 '택시 이용 횟수'가 많으면 높은 것, '택시 이용 횟수'가 적으면 낮은것이다.
따라서 구하고자 하는 것을 '날씨에 따른 택시 이용 횟수 확인'으로 할 수 있다.

우리가 확보한 데이터는 다음과 같다.
* 택시의 운행 데이터와 시간대별 날씨 데이터를 조인한 데이터이다.
* 하나의 운행에 대해 다음과 같은 변수(정보)들이 있다. "출발 시각, 출발 시간대, 출발 날짜, 날씨, 기온"
* 날씨 : 흐림, 맑음, 비옴, 흐리고 비옴 .... 등의 범주형 데이터
* 기온 : 화씨 온도를 나타내는 수치형 데이터
* 출발 날짜 : 연,월,일에 해당하는 범주형 데이터
* 출발 시간대 : 0~1, 1~2, 2~3, ... 23~24, 하루 24시간을 1시간 단위로 나눈 24개의 구간 데이터. 범주형 데이터 이다.

우리는 날씨별로 운행 횟수를 셀 수 있다. 특정 날씨에 운행 횟수가 많으면 그 날씨에 수요가 많은 것이라고 할 수 있을까?
그렇지 않다. 다음과 같은 반례를들 수 있다. 비가 오늘 날씨보다 맑은 날씨에 운행 횟수가 많았다. 하지만 실제로는 비가 오는 날씨에 택시 수요가 많다. 해당 기간 중 맑은 날씨 자체가 많았기 때문에 맑은 날씨의 운행 횟수가 많았던 것이다.

'수요가 많은 것'을 어떻게 나타낼지 종속 변수를 명확하게 잘 정해야 한다. 각 날씨에 해당하는 날짜들의 수를 고르게 맞춘 후 날씨별 택시 운행 횟수를 수집한다든가 각 날씨에 해당하는 일 혹은 시간 당 평균 운행 횟수를 활용하는 방법이 있을 수 있다.

2. 구하고자 하는 것 : 같은 날씨에 해당하는 운행들의 1시간 당 평균 운행 횟수
이것이 바로 위에서 구한 데이터이다.

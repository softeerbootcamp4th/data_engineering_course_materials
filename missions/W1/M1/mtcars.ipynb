```python
print(cars.columns)
cars.rename(columns={'Unnamed: 0': 'name'}, inplace=True)
print(cars.columns)
```

```python
print(cars.info())
```

```python
print(cars.describe())
```

```python
cars.set_index('name', inplace=True)
cars
```

```python
print(cars['gear'].nunique())
print(cars['am'].nunique())
```

```python
cross_tab = pd.crosstab(cars['gear'], cars['am'])
cross_tab
```

```python

a = cross_tab.stack().plot.bar(rot=0, title='# of Cars by Gear x Transmission',
                               xlabel='(Gear, Auto Transmission)',
                               ylabel='# of Cars')
```

```python
cars.pivot_table(index='gear', columns='am', values=[], aggfunc=len, fill_value=0)
```

```python
import matplotlib.pyplot as plt

fig, axes = plt.subplots(nrows=4, ncols=3, figsize=(8, 8))
axes = axes.flatten()
for i, column in enumerate(cars.columns):
    cars[column].hist(ax=axes[i])
    axes[i].set_title(f'{column}')
fig.suptitle("Histograms of Variables", fontsize=16)
plt.tight_layout()
plt.show()
```

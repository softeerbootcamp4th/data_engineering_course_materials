```python
gear_am_cross = pd.crosstab(data["gear"], data["am"])
```

```python
gear_am_cross.plot(kind='bar')
plt.title('# of Cars by Gear x Transmission', fontsize=10)
plt.xlabel('(Gear, Auto Transmission)', fontsize=10)
plt.ylabel('# of Cars', fontsize=10)
```

```python
cnt = dict()
uv_gear = list(map(int, sorted(list(data['gear'].unique()))))
uv_am = list(map(int, sorted(list(data['am'].unique()))))


for val_gear in uv_gear:
    for val_am in uv_am:
        select_gear = data[data['gear']==val_gear]
        result = select_gear[data['am']==val_am]
        cnt[(val_gear, val_am)] = int(result.shape[0])
        
print(f"key: (gear, am), value: number of cars\n {cnt}")
```

```python
label = list(map(str, (list(cnt.keys()))))
val = list(cnt.values())

plt.bar(label, val)
plt.title('# of Cars by Gear x Transmission', fontsize=10)
plt.xlabel('(Gear, Auto Transmission)', fontsize=10)
plt.ylabel('# of Cars', fontsize=10)
```

## car_name
* 차 이름

## mpg
* 연비
* miles per gallon
    * 1 mile = 1.609344 km
    * 1 gallon = 3.785411784 L

## cyl
* 실린더 수

* 실린더
    * 연료와 공기의 혼합물이 연소하여 차량을 움직이는 힘을 생성하는 공간
    * 엔진 블록 내부에 위치하며, 각 실린더에는 피스톤이 들어있다.
    * 피스톤은 실린더 내에서 상하로 움직이며, 이 움직임이 엔진의 동력을 만듭니다.

## disp
* 엔진 배기량
* Displacement
    * 실린더가 한 사이클 동안 흡입, 압축, 폭발, 배기를 통해 이동할 수 있는 총 공기 및 연료 혼합물의 부피를 나타냅니다.
    * 리터(L) 또는 cc 단위
    * 실린더 수 x 실린더 부피
* 중요성
    1. 성능: 일반적으로 더 큰 배기량의 엔진은 더 많은 공기와 연료를 연소시킬 수 있어 더 높은 출력을 낼 수 있다.
    2. 연비: 배기량이 큰 엔진은 더 많은 연료를 소비할 수 있어 연비가 낮을 수 있다.
    3. 세금 및 규제: 일부 국가에서는 차량의 배기량에 따라 세금이나 보험료가 다르게 부과된다.

## hp
* 엔진이나 모터가 내는 출력의 단위를 의미이다. 
* 마력은 원래 한 마리가 1분 동안 수행할 수 있는 일의 양을 기준으로 한 단위이다.

* 중요성
    1. 성능: 마력은 차량의 엔진이 얼마나 강력한지를 나타냅니다. 높은 마력은 더 빠른 가속과 높은 최고 속도를 가능하게 합니다.
    2. 가속도: 높은 마력을 가진 차량은 더 짧은 시간 안에 높은 속도에 도달할 수 있습니다.
    3. 견인력: 마력이 높은 차량은 더 무거운 화물을 견인할 수 있습니다.

## drat
* 차량의 후륜 차축에서 구동축과 바퀴 사이의 기어 비율 = rear axle ratio
* rear axle ratio가 3.73:1 -> 구동축이 한 번 회전할 때 바퀴는 3.73번 회전

* 중요성
    1. 가속 성능: 낮은 기어 비율(예: 3.73:1)은 더 높은 가속력을 제공, 높은 RPM에서 더 많은 연료를 소비할 수 있다.
    2. 연비: 높은 기어 비율(예: 2.73:1)은 더 낮은 RPM에서 더 적은 연료를 소비할 수 있어 연비가 좋아질 수 있다. 그러나 가속 성능은 저하될 수 있다.
    3. 견인 능력: 낮은 기어 비율은 더 높은 토크를 제공하여 견인 능력을 향상시킨다.
* 예시
    1. 낮은 기어 비율(높은 숫자): 4.10:1
        * 더 강력한 가속과 견인력을 제공
        * 고속 주행 시 연비가 떨어질 수 있음
    2. 높은 기어 비율(낮은 숫자): 2.73:1
        * 고속 주행 시 연비가 좋음
        * 가속 성능과 견인력이 상대적으로 낮음

## wt
* Weight (lb/1000)
* (lb/1000) : 파운드를 1000으로 나눈 값을 의미

## qsec
* quarter mile time
    * 차량이 정지 상태에서 출발하여 1/4마일(약 402미터) 거리를 주파하는 데 걸리는 시간
    * 단위 : 초(sec)

## vs
* 엔진의 실린더가 어떻게 배치
* V : 0
    * V형 엔진 (V-shaped engine)
    * 실린더가 V자 형태로 배열된 방식
    * 장점
        * 컴팩트한 설계로 동일한 공간에서 더 많은 실린더를 배치할 수 있다.
        * V6 (V형 6기통), V8 (V형 8기통)
* S : 1
    * 직렬형 엔진 (Straight engine)
    * 실린더가 일직선으로 배열된 방식
    * 장점
        * 설계가 단순하고 제작이 용이
        * I4 (직렬 4기통), I6 (직렬 6기통)

## am
* 변속기(Transmission) 종류
    * 0 : Automatic transmission (자동 변속기)
    * 1 : Manual transmission (수동 변속기)

## gear
* 차량의 전진 기어 수
* 일반적으로 높은 기어 수는 더 좋은 가속성과 고속 주행이 가능하다.
    
* 중요성
    1. 성능: 기어 수가 많을수록 차량은 더 넓은 속도 범위에서 효율적인 가속과 고속 주행을 할 수 있다.
    2. 운전 경험: 운전자는 다양한 상황에서 적절한 기어를 선택하여 운전 편의성과 효율성을 높일 수 있다.
    3. 경주나 스포츠 차량: 고성능 차량은 종종 더 많은 기어를 가지고 있어 가속성과 주행 성능을 극대화한다.

## carb
* 차량이 장착한 캐브레이터(carburetor)의 수
* carburetor는 내연기관에서 연료와 공기를 혼합하여 실린더로 보내주는 장치이다.

* 중요성
    1. 연비 및 성능: 캐브레이터의 수는 연료와 공기의 혼합을 조절하여 엔진의 연비와 성능에 영향을 줄 수 있다.
    2. 유지 보수: 캐브레이터는 조정이 필요할 수 있으며, 여러 개가 있는 경우 유지 관리가 더 복잡할 수 있다.
    3. 역사적인 측면: 오래된 모델의 자동차에서는 캐브레이터가 기본적으로 사용되었으며, 이는 해당 모델의 역사적인 특징이 될 수 있다.

### 각 feature 간 상관계수

```python
mtcars_matrix = without_car.corr()
mtcars_matrix
```

### 상관계수가 0.7 이상은 관계성이 크다는 뜻이다.

```python
threshold = 0.7
corr_pairs = mtcars_matrix.unstack().reset_index()
corr_pairs.columns = ['Row', 'Column', 'Value']

filtered_corr = corr_pairs[(corr_pairs['Value'] >= threshold) & (corr_pairs['Row'] != corr_pairs['Column'])]

filtered_corr = filtered_corr.drop_duplicates(subset=['Value']).reset_index(drop=True)

print(filtered_corr)
```

### 상관계수가 -0.7보다 작다는 것은 반비례 관계성이 크다는 뜻이다.

```python
threshold = -0.7
corr_pairs = mtcars_matrix.unstack().reset_index()
corr_pairs.columns = ['Row', 'Column', 'Value']

filtered_corr = corr_pairs[(corr_pairs['Value'] <= threshold) & (corr_pairs['Row'] != corr_pairs['Column'])]

filtered_corr = filtered_corr.drop_duplicates(subset=['Value']).reset_index(drop=True)

print(filtered_corr)
```

## 상관계수의 절댓값이 0.7보다 큰 Feature의 관계를 보면 Linear한 경향성을 볼 수 있다.

```python
sns.pairplot(mtcars)
plt.suptitle('Pairplot of mtcars dataset', y=1.02)
plt.show()
```

```python
import numpy as np

high_corr_pairs = mtcars_matrix_abs[(mtcars_matrix_abs >= 0.7) & (mtcars_matrix_abs < 1.0)]
high_corr_pairs = high_corr_pairs.stack().index.tolist()

num_pairs = len(high_corr_pairs)
num_cols = 2
num_rows = int(np.ceil(num_pairs / num_cols))

fig, axes = plt.subplots(num_rows, num_cols, figsize=(15, num_rows * 5))
axes = axes.flatten()

for i, (var1, var2) in enumerate(high_corr_pairs):
    sns.scatterplot(data=mtcars, x=var1, y=var2, hue=(mtcars[var1] + mtcars[var2]), palette='coolwarm', edgecolor='k', ax=axes[i])
    sns.regplot(x=var1, y=var2, data=mtcars, scatter=False, color='blue', ax=axes[i])
    axes[i].set_title(f'{var1} vs {var2} (Correlation: {mtcars_matrix_abs.loc[var1, var2]:.2f})')
    axes[i].set_xlabel(var1)
    axes[i].set_ylabel(var2)

for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```

4. 실린더 수- VS(v형이 0, 직렬이 1) -> VS가 0 또는 1이라서 정확한 비교가 어렵다.
5. 배기-후륜차축비 -> 배기가스가 많다는 것은 엔진 연소를 많이 발생시킨다는 말이다. 후륜차축비가 적다는 것은 drive shaft를 적게 돌리고도 후륜을 1바퀴 돌릴 수 있다는 것이다.
6. 배기-VS(v형이 0, 직렬이 1) -> 큰 배기량 엔진은 일반적으로 V형 엔진인 경우가 많다. V형 엔진은 고출력을 내기 위해 설계되며, 직렬 엔진에 비해 더 큰 배기량을 가지는 경향이 있다. 그러므로 반비례적인 상관관계가 있다.
7. 마력-qsec -> 마력이 높을수록 차량은 더 빠르게 가속할 수 있으며, 1/4 마일을 주파하는 데 걸리는 시간이 줄어든다. 따라서 마력이 높아질수록 qsec 값은 낮아져 반비례적인 상관관계를 가진다.
8. 마력-VS(v형이 0, 직렬이 1) -> 마력이 높은 엔진은 일반적으로 V형 엔진인 경우가 많다. V형 엔진은 직렬 엔진보다 높은 출력을 내는 경향이 있으므로, 마력이 높아질수록 vs 값은 낮아져 반비례적인 상관관계를 갖는다.
9. 기어비 (drat)-무게 -> 무거운 차량은 일반적으로 기어비가 낮은 경향이 있다. 무거운 차량은 높은 속도로 주행하기 어렵고, 더 높은 토크를 필요로 하기 때문에 기어비가 낮아지는 경우가 많다. 따라서 중량이 증가할수록 drat 값은 낮아져 반비례적인 상관관계를 갖는다.

# Business Insight
1. 마력이 높을수록 미국 시장에서 유효할 것이다. 왜냐하면 한국과 다르게 미국은 마트를 갈 때도 픽업트럭으로 수 마일을 달려서 마트로 간다. 이때는 마력이 중요하다. 마력을 늘이기 위해 연비를 줄이고 배기가스를 늘이는 trade-off를 신경써야 한다.

2. 유럽 시장은 연비와 배기가스 배출량을 매우 신경 쓴다. 그러므로 유럽 시장에 진출하기 위해서는 연비를 늘리고 배기가스 배출을 줄이는 노력이 필요하다. 기화기는 연료와 공기를 배합하여 혼합기를 만드는 장치이다. 최적의 공연비를 맞추기 위해 기화기를 잘 만들어야하는 목표를 설정할 수 있다.

3. 실린더-배기가스 배출량을 비교하여 피드백 할 수 있다. 실린더 개수에 맞는 배기가스가 잘 배출되는지, 마력이 제대로 출력이 되는지 검증할 수 있다.

4. 연비가 좋은 차량을 개발할 때 상관관계가 높은 파라미터를 우선으로 조정함으로써 개발시간을 단축할 수 있다. 또한 목표를 정확하게 접근할 수 있다.

### 주요 특징
1. 엔진: 2.0리터 직렬 4기통 엔진을 탑재하여 경제적이면서도 충분한 성능을 제공한다.
2. 디자인: 클래식하고 단정한 디자인으로 유명하며, 볼보의 전통적인 스칸디나비아 스타일을 잘 나타낸다.
3. 내구성: 높은 내구성과 안전성을 자랑하는 차량으로, 당시 스웨덴의 안전 기준을 충족시키는 기술적인 특징을 갖추고 있다.
4. 실용성: 넓은 내부 공간과 편안한 승차감을 제공하여 가족용 차량으로 인기가 있었다.
5. 기술 혁신: 당시 최신 기술을 적용하여 경제성과 성능을 균형있게 제공한다.

* Volvo 142E는 볼보의 전통적인 스칸디나비아 디자인과 안전성을 갖춘 중형 스포츠 세단으로, 경제적인 운전과 넓은 내부 공간을 통해 가족용 차량으로서 인기를 끌었다.
* 당시의 기술 혁신과 내구성으로도 주목받는 모델이다.
